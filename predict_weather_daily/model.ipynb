```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping

df = pd.read_csv("new_datasets.csv")

df.dropna(inplace=True)

df["CET"] = pd.to_datetime(df["CET"], format='%Y-%m-%d')

X = df["CET"]
y_temp = df["Mean TemperatureC"]
y_hum = df[" Mean Humidity"]
y_ws = df[" Max Wind SpeedKm/h"]

min_date = X.min()
max_date = X.max()
X_normalized = (X - min_date) / (max_date - min_date)

X_train, X_test, y_temp_train, y_temp_test, y_hum_train, y_hum_test, y_ws_train, y_ws_test = train_test_split(
    X_normalized, y_temp, y_hum, y_ws, test_size=0.2, random_state=42)

model = Sequential()

model.add(Dense(1, input_shape=(1,), activation='linear'))  # Temperature
model.add(Dense(1, activation='linear'))                   # Humidity
model.add(Dense(1, activation='linear'))                   # Wind Speed

model.compile(optimizer=Adam(), loss="mean_absolute_error", metrics=["mean_squared_error"])

early_stop = EarlyStopping(monitor="loss", verbose=0, patience=50, mode="auto")

model.fit(X_train, [y_temp_train, y_hum_train, y_ws_train], epochs=100, validation_split=0.2)

df.to_csv("new_datasets.csv")

# Save the model if needed
# model.save("my_model.h5")

# tarih_obj = pd.to_datetime("2013-06-25")
# zaman_damgasi = (tarih_obj - min_date) / (max_date - min_date)
# tarih_tensor = tf.constant([zaman_damgasi], dtype=tf.float32)
# temp_pred, hum_pred, ws_pred = model.predict(tarih_tensor)
# temp_pred = temp_pred * (y_temp.max() - y_temp.min()) + y_temp.min()
# hum_pred = hum_pred * (y_hum.max() - y_hum.min()) + y_hum.min()
# ws_pred = ws_pred * (y_ws.max() - y_ws.min()) + y_ws.min()

# print(f"Scaklk Tahmini: {temp_pred[0][0]:.2f}C")
# print(f"Nem Tahmini: {hum_pred[0][0]:.2f}%")
# print(f"Rzgar Hz Tahmini: {ws_pred[0][0]:.2f} km/h")
```
